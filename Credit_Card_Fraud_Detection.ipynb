```python
#Encoding the categorical column
df=pd.read_csv("/content/drive/MyDrive/creditcardfraud.csv")
```

```python
df.head()
```

```python
## By using ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
cat_col=df.select_dtypes(include='object').columns
df[cat_col]=oe.fit_transform(df[cat_col])
df.head()
```

```python
for col in df.columns:
  if df[col].dtype=='datetime64[ns]':
    df[col]=df[col].apply(lambda x:x.timestamp())
```

```python
#checking categorical columns

df.info()
```

```python
# Separating the data into target and feature columns.
#seperate target and feature colums
X=df.drop(['is_fraud'],axis=1)#feature columns
y=df['is_fraud']#target column
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
#Scaling the data using StandardScaler
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()
X_scaled=scale.fit_transform(X)
X_scaled_df=pd.DataFrame(X_scaled)
X_scaled_df.head()
```

```python
#splitting our data into training and testing in 70:30 ratio
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Applying various models
# Applying Linear regression

from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
log.score(X_train,y_train)
```

```python
y_pred=log.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
## Applying SVM  model

from sklearn.svm import SVC
from sklearn.svm import LinearSVC

linsvc=LinearSVC(random_state=1)

linsvc.fit(X_train,y_train)
```

```python
#checking training accuracy
linsvc.score(X_train,y_train)
```

```python
## Applying decision tree classifier model.

from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier(criterion='entropy',max_depth=15) #by default criterion for splitting in decision tree is gini index.

dtc.fit(X_train,y_train)
```

```python
#checking the accuracy on training data
dtc.score(X_train,y_train)
```

```python
#checking the accuracy on testing data
dtc.score(X_test,y_test)
```

```python
#Using ensemble methods
#Using RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
rf.score(X_train,y_train)
```

```python
#checking the testing accuracy.
rf.score(X_test,y_test)
```

```python
## Using Adaboost

from sklearn.ensemble import AdaBoostClassifier

ada=AdaBoostClassifier(algorithm='SAMME')
ada.fit(X_train,y_train)
```

```python
ada.score(X_train,y_train)
```

```python
ada.score(X_test,y_test)
```

```python
## Using Gradiant Boosting

from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
gb.score(X_train,y_train)
```

```python
gb.score(X_test,y_test)
```

```python
## Using XG boosting

from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
```

```python
xgb.score(X_train,y_train)
```

```python
xgb.score(X_test,y_test)
```
